```python
# Que:-1 Program to find the largest of three numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Largest number is:", a)
elif b > c:
    print("Largest number is:", b)
else:
    print("Largest number is:", c)
```

```python
#Que:-2 Program to check whether a year is leap year

year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year")
else:
    print(year, "is not a Leap Year")
```

```python
#Que:-3 Program to check vowel or consonant

char = input("Enter a character: ")

if char in ('a', 'e', 'i', 'o', 'u',
          'A', 'E', 'I', 'O', 'U'):
    print(char, "is a Vowel")
else:
    print(char, "is a Consonant")
```

```python
#Que:-4 Program to check divisibility by 5 and 11

num = int(input("Enter a number: "))

if num % 5 == 0 and num % 11 == 0:
    print(num, "is divisible by both 5 and 11")
else:
    print(num, "is not divisible by both 5 and 11")
```

```python
#Que:-5 Program to find sum of first N natural numbers

n = int(input("Enter value of N: "))

sum = 0
i = 1

while i <= n:
    sum = sum + i
    i = i + 1

print("Sum =", sum)
```

```python
#Que:-6 Program to print multiplication table

num = int(input("Enter a number: "))

i = 1

while i <= 10:
    print(num, "x", i, "=", num * i)
    i = i + 1
```

```python
#Que:-7 Pyramid pattern code

for i in range(1, 5):
    print(" " * (4 - i), end="")
    print("*" * (2 * i - 1))
```

```python
#Que:-8 Number pattern program

for i in range(1, 6):
    for j in range(1, i+1):
        print(j, end="")
    print()
```
